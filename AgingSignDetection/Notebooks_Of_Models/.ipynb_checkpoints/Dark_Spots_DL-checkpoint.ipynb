```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

# Read the Data

```python
images=[]
labels=[]
l=['dark_spots_images','no_dark_spots']
for i in l:
    images_per_folder=os.listdir('Datasets/dark_spots_dataset'+'/'+i)
    for j in images_per_folder:
        image=plt.imread('Datasets/dark_spots_dataset/'+i+'/'+j)
        image=cv2.resize(image,(224,224))
        images.append(image)
        labels.append(i)
        #print("Scanning images in folder: "+ i+'/' + j)
    print("Scanning images in folder "+ i)
```

# Extract Face part from the image

```python
from matplotlib import pyplot
from PIL import Image
from numpy import asarray
from mtcnn.mtcnn import MTCNN
 
error=[]
detector = MTCNN()
def extract_face(pixels, required_size=(224, 224)):

    #pixels = pyplot.imread(filename)
#     cv2.imshow('img',pixels)
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
    
    results = detector.detect_faces(pixels)
    
    x1, y1, width, height = results[0]['box']
    x2, y2 = x1 + width, y1 + height
    
    face = pixels[y1:y2, x1:x2]
   
    image = Image.fromarray(face)
    image = image.resize(required_size)
    face_array = asarray(image)
    return face_array
 

# pixels = extract_face('images.jpg')

# pyplot.imshow(pixels)

# pyplot.show()
k=0
final_images=[]
for image in images:
    try:
        pixels = extract_face(image)
        final_images.append(pixels)
        print("Try: "+str(k))
        k+=1
    except:
        error.append(k)
        final_images.append(image)
        print("Exception: "+str(k))
        k+=1
        
        
#final_images=images
```

```python
error=[]
for i in range(len(final_images)):
    if(final_images[i].shape!=(224,224,3)):
        print(i)
        error.append(i)
```

```python
error
```

```python
i=len(error)-1
while(i>=0):
    final_images.pop(error[i])
    labels=np.delete(labels,error[i])
    i-=1
    
```

```python
print(len(labels))
print(len(final_images))
```

```python
final_images=np.array(final_images)
labels=np.array(labels)
```

```python
labels
```

# one hot encoding

```python
for i in range(len(labels)):
    if labels[i]=='no_dark_spots':
        labels[i]=0
    else:
        labels[i]=1
```

```python
labels
```

# Change to Categorical Variable

```python
from tensorflow.keras.utils import to_categorical
target=to_categorical(labels)
target
```

# Train_Test_Split

```python
from sklearn.model_selection import train_test_split
(trainX, testX, trainY, testY) = train_test_split(final_images, target, test_size=0.1,random_state=42) 
```

# Model Training

# Architecture of model

```python
IMG_SIZE=224
input_shape=(224,224,3)
from tensorflow.keras.applications import EfficientNetB0 as Net
from tensorflow.keras import layers
import tensorflow as tf

inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
conv_base = Net(weights="imagenet", include_top=False, input_shape=input_shape)

```

```python
from tensorflow.keras import models
from tensorflow.keras import layers

dropout_rate = 0.2
model = models.Sequential()
model.add(conv_base)
model.add(layers.GlobalMaxPooling2D(name="gap"))

if dropout_rate > 0:
    model.add(layers.Dropout(dropout_rate, name="dropout_out"))

model.add(layers.Dense(2, activation="softmax", name="fc_out"))
```

```python
conv_base.trainable = False
```

# Compile Model

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

# Training the Model

```python
model.fit(trainX,trainY,epochs=15)
```

# Predicting

```python
prd=model.predict(testX)
```

```python
eval=[]
for i in range(len(prd)):
    eval.append(testY[i].argmax())
```

```python
result=[]
for i in range(len(prd)):
    result.append(prd[i].argmax())
```

# Finding Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(eval,result)
```
